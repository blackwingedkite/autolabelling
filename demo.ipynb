```python
import os
import json
import pandas as pd

# 創建一個空的列表來存儲句子
sentences = []

# 設定資料夾路徑
folder_path = 'data'

# 遍歷資料夾中的所有檔案
for filename in os.listdir(folder_path):
    print(filename)
    if filename.endswith('.jsonl'):
        file_path = os.path.join(folder_path, filename)
        
        # 讀取 JSONLines 檔案
        with open(file_path, 'r') as file:
            for line in file:
                entry = json.loads(line)
                
                # 提取每個段落中的句子
                if 'paragraph' in entry:
                    sentences.append(str(entry['paragraph']))
```

```python
import pandas as pd
import openai
import time
import random


def get_openai_response(prompt):
    try:
        response = client.chat.completions.create(
            model="gpt-3.5-turbo",
            messages=[
                {"role": "system", "content": "You are a helpful financial report expert."},
                {"role": "user", "content": prompt}
            ]
        )
        return response.choices[0].message.content.strip()
    except Exception as e:
        print(f"Error in API call: {e}")
        return None

# Initialize results list
results = []
count = 0

# Iterate through each input paragraph
for index, row in data.iterrows():
    paragraph = row['sentence']
    
    # Ask three questions for each paragraph and answer
    prompt_a = f"{question_a}\n\n{paragraph}"
    answer_a = get_openai_response(prompt_a)
    
    prompt_b = f"{question_b}\n\n{answer_a}"
    answer_b = get_openai_response(prompt_b)

    prompt_c = f"{question_c}\n\n{answer_a}"
    answer_c = get_openai_response(prompt_c)
    
    # Add results to the list
    results.append({
        'paragraph': paragraph,
        'answer_a': answer_a,
        'answer_b': answer_b,
        'answer_c': answer_c
    })
    
    # Avoid exceeding API rate limits
    sleep_time = random.randint(1, 5)    
    time.sleep(sleep_time)

    count += 1
    print(count)
    if count % 10 == 0:
        print(f"Processed {count} paragraphs, saving intermediate results.")
        # Convert results to DataFrame
        results_df = pd.DataFrame(results)
        # Save intermediate results to CSV file
        results_df.to_csv(f'output_results_{count}.csv', index=False)

# Save final results to CSV file
results_df = pd.DataFrame(results)
results_df.to_csv('output_results_final.csv', index=False)
print("Processing complete, results saved to output_results_final.csv")
```

```python
processed_data = pd.read_csv("./output_results_final.csv")
processed_data.columns = ['text', 'highlight', 'type', "topic"]

# 定義正則表達式來提取大括弧中的內容
pattern = r'\{([^}]*)\}'

# 僅對column2和column3應用正則表達式
processed_data['type'] = processed_data['type'].str.extract(pattern, expand=False).apply(lambda x: f'{{{x}}}' if pd.notnull(x) else x)
processed_data['topic'] = processed_data['topic'].str.extract(pattern, expand=False).apply(lambda x: f'{{{x}}}' if pd.notnull(x) else x)
patterns_to_remove = [
    r'# RESPONSE #',
    r'#RESPONSE#',
    r'\n',
    r'<response>',
    r'</response>'
]

# 使用str.replace來刪除特定內容
for pattern in patterns_to_remove:
    processed_data['highlight'] = processed_data['highlight'].str.replace(pattern, '', regex=True)
processed_data['text'] = processed_data['text'].str.strip("[']")
```
